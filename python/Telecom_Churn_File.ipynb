```python
import pandas as pd
import numpy as np
```

```python
# File path check kar
df = pd.read_csv('churn-bigml-80.csv')

# First look
print(df.shape)   # Rows x Columns
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
# Count missing values
print(df.isnull().sum())
```

```python
# check duplicates
df.duplicated().sum()
```

###### Column Names Clean (Professional Touch)

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
df.columns
```

###### Missing Values Check 

```python
df.isnull().sum()
```

```python
df['churn'] = df['churn'].map({True:'Yes', False:'No'})
```

```python
df
```

```python
df['customer_value'] = pd.qcut(
    df['total_intl_charge'],
    q=3,
    labels=['Low Value', 'Medium Value', 'High Value']
)
```

```python
df
```

```python
df['tenure_group'] = pd.cut(
    df['account_length'],
    bins=[0, 100, 200, 300, 400, 500, 1000],
    labels=['<100 days', '100-200', '200-300', '300-400', '400-500', '500+']
)
```

```python
df
```

###### Overall Churn Rate

```python
churn_rate = df['churn'].value_counts(normalize=True) * 100
churn_rate
```

###### Churn vs Tenure Group

```python
import seaborn as sns
import matplotlib.pyplot as plt

churn_tenure = pd.crosstab(df['tenure_group'], df['churn'], normalize = 'index') * 100
churn_tenure.plot(kind='bar', figsize=(10,5))

plt.title("CHURN PERCENTAGE BY TENURE GROUP")
plt.ylabel('PERCENTAGE')
plt.xlabel('TENURE GROUP')
plt.show()
```

###### Churn vs Customer Service Calls 

```python
sns.boxplot(x='churn', y='customer_service_calls', data=df)
plt.title('CUSTOMER SERVICE CALLS VS CHURN')
plt.show()
```

###### High-Value Customers & Churn 

```python
pd.crosstab(df['customer_value'], df['churn'], normalize='index') * 100
```

```python
sns.countplot(x='customer_value', hue='churn', data=df)
plt.title('CHRUN BY CUSTOMER VALUE SEGMENT')
plt.show()
```

###### Usage Pattern vs Churn 

```python
sns.boxplot(x='churn', y='total_day_minutes', data=df)
plt.title('DAY MINUTES USAGE VS CHURN')
plt.show()
```

###### Churn rate by customer value (NUMBERS)

```python
churn_by_value = pd.crosstab(
    df['customer_value'],
    df['churn'],
    normalize='index'
) * 100

churn_by_value
```

###### Average charges by churn (REVENUE PROOF)

```python
df.groupby('churn')['total_intl_charge'].mean()
```
